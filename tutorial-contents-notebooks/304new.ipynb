```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
%matplotlib inline

torch.manual_seed(1)
```

```python
x = torch.linspace(-1, 1, 100)
print(x.size())
x = torch.unsqueeze(x, dim=1)
print(x.size())
y = x.pow(2) + 0.2*torch.rand(x.size())
plt.scatter(x, y)
plt.show()
```

```python
net = nn.Sequential(
    nn.Linear(1, 10),
    nn.ReLU(),
    nn.Linear(10, 10),
    nn.ReLU(),
    nn.Linear(10, 1)
)
print(net)
```

```python
import torch.optim as optim
criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.1)

device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
net.to(device)
X_train = x.to(device)
Y_train = y.to(device)
```

```python
def train():
    for t in range(1000):
        net.zero_grad()
        outputs = net(X_train)
        loss = criterion(outputs, Y_train)
        loss.backward()
        optimizer.step()
    
    with torch.no_grad():
        outputs = net(X_train).cpu()
        plt.scatter(x, y)
        plt.plot(x, outputs, 'r-', lw=5)
```

```python
train()
```

```python
def save():
    plt.figure(1, figsize=(10, 3))
    plt.subplot(131)
    plt.title('Net1')
    plt.scatter(x, y)
    with torch.no_grad():
        outputs = net(X_train).cpu()
        plt.plot(x, outputs, 'r-', lw=5)
    torch.save(net, 'net.pkl')
    torch.save(net.state_dict(), 'net_params.pkl')
```

```python
def restore_net():
    # net2 = torch.load('net.pkl') # 这样加载时device和save时保持一致
    net2 = torch.load('net.pkl', map_location='cpu')
    # print(net2)
    # print(net2.device)
    plt.subplot(132)
    plt.title('Net2')
    plt.scatter(x, y)
    with torch.no_grad():
        outputs = net2(x)
        plt.plot(x, outputs, 'r-', lw=5)
```

```python
def restore_params():
    net3 = nn.Sequential(
        nn.Linear(1, 10),
        nn.ReLU(),
        nn.Linear(10, 10),
        nn.ReLU(),
        nn.Linear(10, 1)
    )
    net3.load_state_dict(torch.load('net_params.pkl')) # 这样加载时device为net3定义的device
    plt.subplot(133)
    plt.title('Net3')
    plt.scatter(x, y)
    with torch.no_grad():
        outputs = net3(x)
        plt.plot(x, outputs, 'r-', lw=5)
```

```python
save()
restore_net()
restore_params()
plt.show()
```
