```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
%matplotlib inline

torch.manual_seed(1)
EPOCH = 20
BATCH_SIZE = 64
LR = 0.005
N_TEST_IMG = 5
DEVICE = 'cuda:0' if torch.cuda.is_available() else 'cpu'
```

```python
transform = transforms.ToTensor()
train_set = torchvision.datasets.MNIST(root='./mnist', train=True, transform=transform, download=True)
train_loader = torch.utils.data.DataLoader(train_set, shuffle=True, batch_size=BATCH_SIZE, num_workers=32)
test_set = torchvision.datasets.MNIST(root='./mnist', train=False, transform=transform, download=True)
test_loader = torch.utils.data.DataLoader(test_set, shuffle=True, batch_size=BATCH_SIZE, num_workers=32)
print(train_set.data.size())
print(train_set.targets.size())
print(test_set.data.size())
print(test_set.targets.size())
plt.imshow(train_set.data[2], cmap='gray')
plt.title('%i' % train_set.targets[2])
plt.show()
```

```python
class AutoEncoder(nn.Module):
    def __init__(self):
        super(AutoEncoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 128),
            nn.Tanh(),
            nn.Linear(128, 64),
            nn.Tanh(),
            nn.Linear(64, 12),
            nn.Tanh(),
            nn.Linear(12, 3)
        )
        self.decoder = nn.Sequential(
            nn.Linear(3, 12),
            nn.Tanh(),
            nn.Linear(12, 64),
            nn.Tanh(),
            nn.Linear(64, 128),
            nn.Tanh(),
            nn.Linear(128, 28*28),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return encoded, decoded
```

```python
net = AutoEncoder()
print(net)

net.to(DEVICE)
criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=LR)

view_data = test_set.data[:N_TEST_IMG].view(-1, 28*28).type(torch.FloatTensor)/255.0

for epoch in range(EPOCH):
    running_loss = 0.0
    for t, (inputs, _) in enumerate(train_loader):
        inputs = inputs.view(-1, 28*28).to(DEVICE)
        net.zero_grad()
        _, outputs = net(inputs)
        loss = criterion(outputs, inputs)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        if t % 400 == 399:
            print('epoch %d step %d  train loss : %.3f' % (epoch+1, t+1, running_loss/400.0))
            running_loss = 0.0
            
            with torch.no_grad():
                _, outputs = net(view_data.to(DEVICE))
                outputs = outputs.cpu()
                f, a = plt.subplots(2, N_TEST_IMG, figsize=(5, 2))
                for i in range(N_TEST_IMG):
                    a[0][i].imshow(view_data[i].view(28, 28), cmap='gray')
                    a[0][i].set_xticks(())
                    a[0][i].set_yticks(())
                for i in range(N_TEST_IMG):
                    a[1][i].clear()
                    a[1][i].imshow(outputs[i].view(28, 28), cmap='gray')
                    a[1][i].set_xticks(())
                    a[1][i].set_yticks(())
                plt.show()
                plt.pause(0.05)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# visualize in 3D plot
view_data = train_set.data[:200].view(-1, 28*28).type(torch.FloatTensor)/255.0
with torch.no_grad():
    encoded_data, _ = net(view_data.to(DEVICE))
    encoded_data = encoded_data.cpu()
    fig = plt.figure(2); ax = Axes3D(fig)
    X, Y, Z = encoded_data[:, 0], encoded_data[:, 1], encoded_data[:, 2]
    values = train_set.targets[:200]
    for x, y, z, s in zip(X, Y, Z, values):
        c = cm.rainbow(int(255*s/9)); ax.text(x, y, z, s.item(), backgroundcolor=c)
    ax.set_xlim(X.min(), X.max()); ax.set_ylim(Y.min(), Y.max()); ax.set_zlim(Z.min(), Z.max())
    plt.show()
```
