```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

torch.manual_seed(1)
```

```python
BATCH_SIZE = 64
LR_G = 0.0001
LR_D = 0.0001
N_IDEAS = 5    # 生成网络的输入size
ART_COMPONENTS = 15    # 生成的样本size
PAINT_POINTS = np.vstack([np.linspace(-1, 1, ART_COMPONENTS) for _ in range(BATCH_SIZE)])
DEVICE = 'cuda:0' if torch.cuda.is_available() else 'cpu'
```

```python
print(PAINT_POINTS[0])
plt.plot(PAINT_POINTS[0], 2 * PAINT_POINTS[0] ** 2 + 1, c='#74BCFF', lw=3, label='upper bound')
plt.plot(PAINT_POINTS[0], 1 * PAINT_POINTS[0] ** 2 + 0, c='#FF9359', lw=3, label='lower bound')
plt.legend(loc='upper right')
plt.show()
```

```python
def artist_works():
    #a = torch.random.uniform(1, 2, size=BATCH_SIZE)[:, np.newaxis]
    a = torch.Tensor(BATCH_SIZE, 1).uniform_(1, 2)
    a = a * (PAINT_POINTS ** 2) + (a - 1)
    return a.float()


artist_works()

G_net = nn.Sequential(
    nn.Linear(N_IDEAS, 128),
    nn.ReLU(),
    nn.Linear(128, ART_COMPONENTS)
).to(DEVICE)

D_net = nn.Sequential(
    nn.Linear(ART_COMPONENTS, 128),
    nn.ReLU(),
    nn.Linear(128, 1),
    nn.Sigmoid()
).to(DEVICE)
```

```python
G_optim = optim.Adam(G_net.parameters(), lr=LR_G)
D_optim = optim.Adam(D_net.parameters(), lr=LR_D)
```

```python
for t in range(10000):
    G_net.zero_grad()
    D_net.zero_grad()
    artist_paintings = artist_works()
    G_ideas = torch.randn(BATCH_SIZE, N_IDEAS)
    G_paintings = G_net(G_ideas.to(DEVICE))
    # print(artist_paintings.size())
    # print(G_paintings.size())
    
    prob_arts = D_net(artist_paintings.to(DEVICE))
    prob_fake = D_net(G_paintings)
    
    D_loss = - torch.mean(torch.log(prob_arts) + torch.log(1.0 - prob_fake))
    G_loss = torch.mean(torch.log(1.0 - prob_fake))
    
    D_loss.backward(retain_graph=True)
    D_optim.step()
    G_loss.backward()
    G_optim.step()
    
    if t % 1000 == 999:
        plt.cla()
        plt.plot(PAINT_POINTS[0], G_paintings.cpu().data.numpy()[0], c='#4AD631', lw=3, label='Generated painting',)
        plt.plot(PAINT_POINTS[0], 2 * np.power(PAINT_POINTS[0], 2) + 1, c='#74BCFF', lw=3, label='upper bound')
        plt.plot(PAINT_POINTS[0], 1 * np.power(PAINT_POINTS[0], 2) + 0, c='#FF9359', lw=3, label='lower bound')
        plt.text(-.5, 2.3, 'D accuracy=%.2f (0.5 for D to converge)' % prob_arts.cpu().data.numpy().mean(), fontdict={'size': 15})
        plt.text(-.5, 2, 'D score= %.2f (-1.38 for G to converge)' % -D_loss.cpu().data.numpy(), fontdict={'size': 15})
        plt.ylim((0, 3));plt.legend(loc='upper right', fontsize=12);plt.draw();plt.pause(0.01)
        plt.show()
```
