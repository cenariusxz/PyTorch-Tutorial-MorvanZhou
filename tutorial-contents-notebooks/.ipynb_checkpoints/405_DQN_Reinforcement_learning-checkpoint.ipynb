```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F
import numpy as np
import gym
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display
```

```python
# Hyper Parameters
BATCH_SIZE = 32
LR = 0.01                   # learning rate
EPSILON = 0.9               # greedy policy
GAMMA = 0.9                 # reward discount
TARGET_REPLACE_ITER = 100   # target update frequency
MEMORY_CAPACITY = 2000
env = gym.make('CartPole-v0')
env = env.unwrapped
N_ACTIONS = env.action_space.n
N_STATES = env.observation_space.shape[0]
```

```python
class Net(nn.Module):
    def __init__(self, ):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(N_STATES, 10)
        self.fc1.weight.data.normal_(0, 0.1)   # initialization
        self.out = nn.Linear(10, N_ACTIONS)
        self.out.weight.data.normal_(0, 0.1)   # initialization

    def forward(self, x):
        x = self.fc1(x)
        x = F.relu(x)
        actions_value = self.out(x)
        return actions_value
```

```python
class DQN(object):
    def __init__(self):
        self.eval_net, self.target_net = Net(), Net()

        self.learn_step_counter = 0                                     # for target updating
        self.memory_counter = 0                                         # for storing memory
        self.memory = np.zeros((MEMORY_CAPACITY, N_STATES * 2 + 2))     # initialize memory
        self.optimizer = torch.optim.Adam(self.eval_net.parameters(), lr=LR)
        self.loss_func = nn.MSELoss()

    def choose_action(self, x):
        x = Variable(torch.unsqueeze(torch.FloatTensor(x), 0))
        #print('x_size:', x.size())
        # input only one sample
        if np.random.uniform() < EPSILON:   # greedy
            actions_value = self.eval_net.forward(x)
            #print('actions_value_size:', actions_value.size())
            #print('actions_value:', actions_value)
            action = torch.max(actions_value, 1)[1].data.numpy()[0].item()     # return the argmax
            #print('actions:', actions)
        else:   # random
            action = np.random.randint(0, N_ACTIONS)
        return action

    def store_transition(self, s, a, r, s_):
        transition = np.hstack((s, a, r, s_))
        # replace the old memory with new memory
        index = self.memory_counter % MEMORY_CAPACITY
        self.memory[index, :] = transition
        self.memory_counter += 1

    def learn(self):
        # target parameter update
        if self.learn_step_counter % TARGET_REPLACE_ITER == 0:
            self.target_net.load_state_dict(self.eval_net.state_dict())
        self.learn_step_counter += 1

        # sample batch transitions
        sample_index = np.random.choice(MEMORY_CAPACITY, BATCH_SIZE)     # 在 0 - MEMORY_CAPACITY-1 内随机选 BATCH_SIZE 个数，作为用于更新的样本编号
        b_memory = self.memory[sample_index, :]    # BATCH_SIZE * ( 4 + 1 + 1 + 4)
        b_s = Variable(torch.FloatTensor(b_memory[:, :N_STATES]))   # BATCH_SIZE * 4
        b_a = Variable(torch.LongTensor(b_memory[:, N_STATES:N_STATES+1].astype(int)))  # BATCH_SIZE * 1
        b_r = Variable(torch.FloatTensor(b_memory[:, N_STATES+1:N_STATES+2]))   # BATCH_SIZE * 1
        b_s_ = Variable(torch.FloatTensor(b_memory[:, -N_STATES:]))     # BATCH_SIZE * 4

        # q_eval w.r.t the action in experience
        self.eval_net.zero_grad()
        q_eval = self.eval_net(b_s)
        #print('before size:', q_eval.size())
        q_eval = q_eval.gather(1, b_a)  # shape (batch, 1)
        #print('after size:', q_eval.size())
        q_next = self.target_net(b_s_).detach()     # detach from graph, don't backpropagate
        #print('q_next_max:', q_next.max(1))
        #print('q_next_max[0]:', q_next.max(1)[0])
        #print('GAMMA * q_next_max[0]:', GAMMA * q_next.max(1)[0])
        #print('br:', b_r)
        #print('GAMMA * q_next_max[0]_size:', torch.unsqueeze(GAMMA * q_next.max(1)[0], dim=1).size())
        #print('br_size:', b_r.size())
        q_target = b_r + torch.unsqueeze(GAMMA * q_next.max(1)[0], dim=1)   # shape (batch, 1)
        #print('!!!!!!!!', q_eval.size())
        #print('!!!!!!!!', q_target.size())
        
        loss = self.loss_func(q_eval, q_target)
        loss.backward()
        self.optimizer.step()
```

```python
dqn = DQN()
```

```python
def show_state(env, step=0, info=""):
    plt.figure(3)
    plt.clf()
    plt.imshow(env.render(mode='rgb_array'))
    plt.title("Step: %d %s" % (step, info))
    plt.axis('off')

    display.clear_output(wait=True)
    display.display(plt.gcf())

print('\nCollecting experience...')
for i_episode in range(400):
    s = env.reset()
    ep_r = 0
    while True:
        env.render()
        # show_state(env, 0, 'Ep: %d | Ep_r: %.2f' % (i_episode+1, round(ep_r, 2)))
        a = dqn.choose_action(s) # 0 or 1 应该是向左还是向右

        # take action
        s_, r, done, info = env.step(a)    # s 状态  r 环境给定奖赏  done 情节是否结束  info 暂无

        # modify the reward
        x, x_dot, theta, theta_dot = s_    # s_ : numpy size (4)    表示状态
        r1 = (env.x_threshold - abs(x)) / env.x_threshold - 0.8
        r2 = (env.theta_threshold_radians - abs(theta)) / env.theta_threshold_radians - 0.5
        r = r1 + r2

        dqn.store_transition(s, a, r, s_)    # s ndarray(4)   a 0/1    r float   s_ ndarray(4)

        ep_r += r
        if dqn.memory_counter > MEMORY_CAPACITY:
            dqn.learn()
            if done:
                print('Ep: ', i_episode+1,
                      '| Ep_r: ', round(ep_r, 2))

        if done:
            break
        s = s_
```
